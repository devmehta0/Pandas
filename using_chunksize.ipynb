```python
concator_df = pd.DataFrame(columns=['event_type','size'])
print(concator_df)
```

```python
# this for loop will iterate for each chunk/group of 10,000 rows and counts number of events for each types
for df_ in df:
    df3 = df_.groupby('event_type', as_index = False).size()
    concator_df = pd.concat([concator_df,df3])
```
