### Sorting index and values

```python
df.sort_index(ascending=True)
```

```python
# setting "event_time" as index. 

df_index = df.set_index('user_id') 
df_index.sort_index(inplace = True)
```

```python
# You can see in below result, index got changed to user_id.
df_index
```

```python
# sort_values() : sorts the dataframe based the column name specify in 'by' parameter

df_index.sort_values(by='product_id', ascending=True)
```

```python
# Similar to 'ORDER BY event_time, event_type' in sql. Can change ascending parameter as per required order ASC/DESC as True/False

df_index.sort_values(by=['user_id','product_id'], ascending=[True,True])
```

### Filling Null Values

```python
# fillna(): fills the null value with provided data.
df2 = df.fillna('unknown')
```

```python
df2.isnull().sum()
```

```python
# filling different value for different columns

df3 = df.fillna({
    'category_code': 'unknown category',
    'brand' : 'unknown brand'
})
```

```python
df3
```

```python
df3.isnull().sum()
```

### Dropping NULL values

```python
df
```

```python
df4 = df.dropna()
```

```python
df4
```

<li>You can see in above query, the decrease in number of rows from 10,000 to 6172, which are deleted due to null values.

```python
# Determine if row or column is removed from DataFrame, when we have
#     at least one NA or all NA.
# how parameter has two options:
# - ALL : row will be dropped when all the values in row are NULL
# - ANY : row will be dropped when any of the value in row is NULL

df.dropna(how= 'all')
```

<li>You can see in above result that no rows got dropped as we have used ALL in 'how' parameter.
